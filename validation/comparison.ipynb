```python
xgb_win = []
nn1_win = []
nn2_win = []
knn_win = []
tie = []
for metric in metrics:
    xgb_cnt = 0
    nn1_cnt = 0
    nn2_cnt = 0
    knn_cnt = 0
    tie_cnt = 0
    for (r1,r2,r3,r4) in zip(xgb_kappa_metric[metric],nn1_kappa_metric[metric],nn2_kappa_metric[metric],knn_kappa_metric[metric]):
        if max(r1,r2,r3,r4) == r1:
            xgb_cnt += 1
        elif max(r1,r2,r3,r4) == r2:
            nn1_cnt += 1
        elif max(r1,r2,r3,r4) == r3:
            nn2_cnt += 1
        elif max(r1,r2,r3,r4) == r4:
            knn_cnt += 1
        else:
            tie_cnt += 1
    xgb_win.append(xgb_cnt)
    nn1_win.append(nn1_cnt)
    nn2_win.append(nn2_cnt)
    knn_win.append(knn_cnt)
    tie.append(tie_cnt)
```

```python
df_kappa = pd.DataFrame(columns=["XGB","NN1","NN2","KNN","Tie"],index=metrics)
```

```python
df_kappa['XGB'] = xgb_win
df_kappa['NN1'] = nn1_win
df_kappa['NN2'] = nn2_win
df_kappa['KNN'] = knn_win
df_kappa['Tie'] = tie
```

```python
def result_transform(df):
    df_grouped = df.iloc[:,1:].groupby('Classifier').agg(['mean', 'std'])
    # Define a function to format the mean and std deviation
    def format_mean_std(row):
        return f"{row['mean']:.3f}({row['std']:.3f})"

    # Apply this function to each metric
    for col in df.columns[2:]:  # Skip 'Test_dataset' and 'Classifier' columns
        df_grouped[(col, 'Formatted')] = df_grouped[col].apply(format_mean_std, axis=1)

    # Now we drop the original mean and std columns as we have the formatted column
    df_formatted = df_grouped.drop(columns=[(col, 'mean') for col in df.columns[2:]] + 
                                        [(col, 'std') for col in df.columns[2:]])

    # # Flatten the multi-level column names
    df_formatted.columns = [' '.join(col).strip() if 'Formatted' not in col else ' '.join(col).split()[0] 
                            for col in df_formatted.columns.values]

    df_formatted.reset_index()

    return df_formatted
```
