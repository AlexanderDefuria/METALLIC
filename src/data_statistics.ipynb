```python
import os
import pandas as pd
import numpy as np
```

```python
source_directory = './Dataset/'


rows_list = []
features_list = []
classes_list = []
minority_cases_list = []
imbalance_ratio_list = []


for filename in os.listdir(source_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(source_directory, filename)

        
        df = pd.read_csv(file_path)

        
        rows_list.append(len(df))
        features_list.append(len(df.columns) - 1)  
        num_classes = len(df.iloc[:, -1].unique())
        classes_list.append(num_classes)

        class_counts = df.iloc[:, -1].value_counts()
        minority_cases_list.append(class_counts.min())
        imbalance_ratio_list.append(class_counts.max() / class_counts.min())

        if num_classes == 2:
            if class_counts.index[0] > class_counts.index[1]:
                df.iloc[:, -1] = df.iloc[:, -1].map({class_counts.index[0]: 0, class_counts.index[1]: 1})
        else:         
            new_labels = {cls: i for i, cls in enumerate(class_counts.index)}
            df.iloc[:, -1] = df.iloc[:, -1].map(new_labels)

        
        df.to_csv(file_path, index=False)
        print(filename)  
        print(df.iloc[:, -1].value_counts())


stats_df = pd.DataFrame({
    "Rows": rows_list,
    "Features": features_list,
    "Classes": classes_list,
    "Minority Cases": minority_cases_list,
    "Imbalance Ratio": imbalance_ratio_list
})


overall_stats = pd.DataFrame({
    "Min": stats_df.min(),
    "Max": stats_df.max(),
    "Median": stats_df.median(),
    "Mean": stats_df.mean(),
    "Std": stats_df.std()
})
```

```python
print("Overall Statistics:")
print(overall_stats)
```
